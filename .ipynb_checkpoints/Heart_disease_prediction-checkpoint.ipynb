### Checking correlation between columns

```python
# Correlation Heatmap
corr_matrix = dataset.corr()
plt.figure(figsize = (10,8))
plt.title("Correlation Heatmap of Heart Disease Dataset")
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.show();
```
